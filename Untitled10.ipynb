```python
%%writefile vector.cu

#include<cuda_runtime.h>
#include<iostream>
#include<stdlib.h>
#include<chrono>

using namespace std;
using namespace std::chrono;

#define N 100000

__global__
void add(int *A, int *B, int *C, int n)
{
    int tid=blockIdx.x*blockDim.x+threadIdx.x;
    if(tid<n){
        C[tid]=A[tid]+B[tid];
    }
}

void sequentialAdd(int *A, int *B, int *C, int n)
{
    for(int i = 0; i < n; i++)
    {
        C[i] = A[i] + B[i];
    }
}


void initialize(int *vector,int n){
  for(int i=0;i<n;i++){
    vector[i]=rand()%100;
  }
}


void Print(int *vector,int n){
  cout<<"Starting 10"<<endl;

  for(int i=0;i<10;i++){
    cout<<vector[i]<<" ";
  }
  cout<<endl;
  cout<<"Ending 10"<<endl;

  for(int i=n-10;i<n;i++){
    cout<<vector[i]<<" ";
  }
  cout<<endl;
}

int main(){
  int *A, *B, *C_seq, *C_parallel;

    size_t bytes = N * sizeof(int);

    // Host Memory
    A = new int[N];
    B = new int[N];
    C_seq = new int[N];
    C_parallel = new int[N];

    initialize(A, N);
    initialize(B, N);

    Print(A,10);
    Print(B,10);

    auto start= high_resolution_clock::now();
    sequentialAdd(A,B,C_seq,N);
    auto end =high_resolution_clock::now();

    auto cpu_duration= duration_cast<microseconds>(end-start);
    cout<<"Time taken by CPU: "<<cpu_duration.count()<<" microseconds"<<endl;

      int *d_A,*d_B,*d_C;
      cudaMalloc(&d_A,bytes);
      cudaMalloc(&d_B,bytes);
      cudaMalloc(&d_C,bytes);

      cudaEvent_t start_t,end_t;
      cudaEventCreate(&start_t);
      cudaEventCreate(&end_t);

      cudaMemcpy(d_A,A,bytes,cudaMemcpyHostToDevice);
      cudaMemcpy(d_B,B,bytes,cudaMemcpyHostToDevice);

       int threads = 256;
        int blocks = (N + threads - 1) / threads;

      cudaEventRecord(start_t);

      add<<<blocks,threads>>>(d_A,d_B,d_C,N);

      cudaEventRecord(end_t);
      cudaEventSynchronize(end_t);


        float GPU_TIme =0;

      cudaEventElapsedTime(&GPU_TIme,start_t,end_t);
      cout<<"Time taken by GPU: "<<GPU_TIme<<" microseconds"<<endl;

      cudaMemcpy(C_parallel,d_C,bytes,cudaMemcpyDeviceToHost);

      cout << "Parallel GPU Time: "
         << GPU_TIme
         << " ms" << endl;

    Print(C_parallel,10);







}
```

```python
%%writefile multiplication.cu
#include<cuda_runtime.h>
#include<iostream>
#include<stdlib.h>
#include<chrono>
#define N 512

using namespace std;
using namespace std::chrono;

__global__
void multiply(int *A,int *B,int *C,int n){
  int row=blockIdx.y*blockDim.y+threadIdx.y;
  int col=blockIdx.x*blockDim.x+threadIdx.x;
  if(row<n && col<n){
    int sum=0;
    for(int k=0;k<n;k++){
      sum+=A[row*n+k] * B[k*n+col];
    }

    C[row*n+col]=sum;
  }
}

void initialize(int *vector,int n){
  for(int i=0;i<N*N;i++){
    vector[i]=rand()%100;
  }
}


void sequentialAdd(int *A, int *B, int *C, int n)
{

  for(int i=0;i<n;i++){
    for(int j=0;j<n;j++){
      int sum=0;
      for(int k=0;k<n;k++){
        sum+=A[i*n+k] * B[k*n+j];
      }

      C[i*n+j]=sum;
    }
  }

}
void print(int *vector,int n){
  for(int i=0;i<n;i++){
    for(int j=0;j<n;j++){
      cout<<vector[i*n+j]<<" ";
    }
    cout<<endl;
  }
}

int main(){
  int size=N*N;
  size_t bytes=size*sizeof(int);
  int *A, *B, *C_seq, *C_parallel;
  A=new int[size];
  B=new int[size];
  C_seq=new int[size];
  C_parallel=new int[size];

  initialize(A,N);
  initialize(B,N);

  cout<<"Print A"<<endl;
  print(A,4);
  cout<<"Print B"<<endl;
  print(B,4);

  auto start= high_resolution_clock::now();
  sequentialAdd(A,B,C_seq,N);
  auto end =high_resolution_clock::now();

  auto cpu_time=duration_cast<microseconds>(end-start);
  cout<<"Time taken by CPU: "<<cpu_time.count()<<" microseconds"<<endl;

  int*d_A,*d_B,*d_C;
  cudaMalloc(&d_A,bytes);
  cudaMalloc(&d_B,bytes);
  cudaMalloc(&d_C,bytes);

  cudaMemcpy(d_A,A,bytes,cudaMemcpyHostToDevice);
  cudaMemcpy(d_B,B,bytes,cudaMemcpyHostToDevice);

  cudaEvent_t start_t,end_t;
  cudaEventCreate(&start_t);
  cudaEventCreate(&end_t);

  dim3 thread(16,16);
  dim3 blocks((N+thread.x-1)/thread.x,(N+thread.y-1)/thread.y);

  cudaEventRecord(start_t);

  multiply<<<blocks,thread>>>(d_A,d_B,d_C,N);

  cudaEventRecord(end_t);
  cudaEventSynchronize(end_t);

  float GPU_TIme =0;

  cudaEventElapsedTime(&GPU_TIme,start_t,end_t);
  cout<<"Time taken by GPU: "<<GPU_TIme<<" microseconds"<<endl;

  cudaMemcpy(C_parallel,d_C,bytes,cudaMemcpyDeviceToHost);

  cout << "Parallel GPU Time: "
         << GPU_TIme
         << " ms" << endl;

  for(int i = 0; i < 4; i++)
    {
        for(int j = 0; j < 4; j++)
        {
            cout << C_parallel[i * N + j] << " ";
        }

        cout << endl;
    }



}
```
